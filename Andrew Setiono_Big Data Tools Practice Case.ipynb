# Answering Question no. 1
**Question:** the store owner wanted to know the **number of film of each genre** and the **average replacement cost of each film by ratings.**

To answer these questions we have to **merge** the film, film_category, and category_id datasets.

```python
#Loading the dataset from the Google Cloud Storage
from google.cloud import storage
film = pd.read_csv('gs://dvdrental01/film.csv', sep=';')
fcateg = pd.read_csv('gs://dvdrental01/film_category.csv', sep=';')
categ = pd.read_csv('gs://dvdrental01/category.csv', sep=';')
```

## Data Cleaning

Before proceeding further, I want to do some data processing to check whether there are missing values, duplicates, or anomalies.

```python
#Create a function to get the number of missing values and duplicates.
def rstr(df, pred=None): 
    obs = df.shape[0]
    types = df.dtypes
    counts = df.apply(lambda x: x.count())
    nulls = df.apply(lambda x: x.isnull().sum())
    distincts = df.apply(lambda x: x.unique().shape[0])
    duplicates = df.duplicated().value_counts()
    missing_ration = (df.isnull().sum()/obs) * 100
    print('Data shape:', df.shape)
    
    cols = ['types', 'counts', 'distincts', 'nulls', 'missing ration', 'duplicates']
    str = pd.concat([types, counts, distincts, nulls, missing_ration, duplicates], axis = 1, sort=True)
    
    str.columns = cols
    return str

#Display missing values and duplicates for each datasets
display(rstr(film))
display(rstr(fcateg))
display(rstr(categ))
```

Using the quick function, we can conclude that there are no missing values or duplicates from each datasets.

## Data Analysis

```python
#Create a dataframe with necessary information
film = film[['film_id','title','replacement_cost','rating']]

#Join the necessary Table
merge_fcateg = pd.merge(fcateg, categ, how="inner", on="category_id")
fcateg_all = pd.merge(film, merge_fcateg, how="inner", on="film_id")
print('Data shape:', fcateg_all.shape)
display(fcateg_all.head())
```

From the dataframe, we find that there are multiple columns of **last_update**. These data possibly came from the film category and category dataframe. But, since I am only interested in finding **the number of films in each genre** and **average replacement cost by rating**, I will ignore both of those columns.

### The number of films in each genre

```python
#Create dataframe for Number of Films in each genre
num_genre = fcateg_all.groupby(['name'])['replacement_cost'].count()\
                                                            .reset_index(name='Number of Films')\
                                                            .sort_values('Number of Films', ascending = False)
num_genre = num_genre.rename(columns={'name':'Genre'})
display(num_genre)

#Visualize the data
plt.figure(figsize=(10, 8))
b = sns.barplot(x='Number of Films', y='Genre', data = num_genre)
b.set_title("Number of Films for each Genre", size = 14)
b.set_xlabel('Number of Films', size = 14)
b.set_ylabel('Genre', size = 14)
```

From the barplot, we can conclude that currently the DVD Rental Store mostly have films from **Sports** genre.

### Average replacement cost by rating

```python
#Check for outliers
sns.catplot(x='rating', y='replacement_cost', data=fcateg_all, kind='box')
```

From the boxplot, we can conclude that there are no outliers in the replacement cost for each ratings.

```python
#Joining the necessary dataframe
avg_cost = pd.DataFrame(fcateg_all.groupby('rating')[['replacement_cost']].mean().\
                        sort_values(by='replacement_cost', ascending = False).reset_index())
display(avg_cost)

#Visualizing the data
plt.figure(figsize=(10, 8))
b = sns.barplot(x='rating', y='replacement_cost', data = avg_cost)
b.set_title("Average Replacement Cost by Ratings", size = 14)
b.set_xlabel('Average Replacement Cost', size = 14)
b.set_ylabel('Ratings', size = 14)
```

From the barplot, we can conclude that film rated with **PG** have the **lowest average replacement cost**. Meanwhile, the highest average replacement cost came from **PG-13 films**. Nevertheless, there seems to be no huge difference of average replacement cost between each rating. Further analysis need to be conducted to check this.

# Answering Question no. 2 and 3

**Questions:** 
<ol><li> Find out the top 5 countries based on total amount of payment. </li>
<li> Visualize the distribution of sales by genre within the top 1 country in sales. </li></ol>

To answer these questions we have to **merge** the film category ('fcateg_all') from question number 1 with the addition of datasets of (a) inventory, (b) rental, (c) payment, (d) customer, (e) address, (f) city, (g) country

## Data Loading and Cleaning

```python
#Loading other datasets
inv = pd.read_csv('gs://dvdrental01/inventory.csv', sep=';')
rent = pd.read_csv('gs://dvdrental01/rental.csv', sep=';')
pay = pd.read_csv('gs://dvdrental01/payment.csv', sep=';')
cust = pd.read_csv('gs://dvdrental01/customer.csv', sep=';')
addr = pd.read_csv('gs://dvdrental01/address.csv', sep=';')
city = pd.read_csv('gs://dvdrental01/city.csv', sep=';')
country = pd.read_csv('gs://dvdrental01/country.csv', sep=';')

#Check for duplicates and null values
dataset = [inv, rent, pay, cust, addr, city, country]
for i in dataset:
    display(rstr(i))
```

Based on the **Table Summary above**, we can see that there are some data missing in ‘rent’ and ‘addr’ datasets. 

In the ‘rent’ dataset, there are missing data in return_date column. It is possible that the missing values are because the film is still being rent. Thus, I think it is reasonable to keep the data.

In the ‘addr’ dataset, there are missing data in the address2, district, phone number, column, and postal code. Nevertheless, I am only interested in knowing the amount of payment for each country. And, since it might be reasonable that these data are not required when the customers administer their rent, I decide to also keep the data as it is.

## Joining Datasets

```python
#Choose only the necessary info from each datasets.
fcat = fcateg_all[['film_id', 'category_id', 'name']]
inv2 = inv[['film_id','inventory_id']]
rent2 = rent[['inventory_id','rental_id','customer_id']]
pay2 = pay[['customer_id','payment_id', 'amount']]
cust2 = cust[['customer_id','address_id']]
addr2 = addr[['address_id','city_id']]
city2 = city[['city_id','country_id']]

#Joining the datasets
a = pd.merge(fcat, inv2, how='inner', on='film_id')
b = pd.merge(a, rent2, how='inner', on='inventory_id')
c = pd.merge(b, pay2, how='inner', on='customer_id')
d = pd.merge(c, cust2, how='inner', on='customer_id')
e = pd.merge(d, addr2, how='inner', on='address_id')
f = pd.merge(e, city2, how='inner', on='city_id')
df_all = pd.merge(f, country, how='inner', on='country_id')

print('Data shape:', df_all.shape)
print('Data shape after duplicate dropped:', display(df_all.drop_duplicates()))
print('Data shape after null value dropped:', display(df_all.dropna()))
```

From the results above, we can conclude that the joining process have been successful. The overall dataset has **no duplicates** or **missing values** after being joined. 

```python
df_all.info()
```

Here we can see that the data types are pretty much correct. **Country** names are in **string** and the **amount of payments** are in **float64** data types. The rest of **id** data are in integer.

## Answer for Questions no. 2
**Find out the top 5 countries based on total amount of payment**

```python
#Create a new dataframe that calculate sales from each countries.
pay_country = df_all.groupby(['country','country_id'])['amount'].sum().reset_index(name='Total Amount of Payment')\
                                                        .sort_values('Total Amount of Payment', ascending = False)\
                                                        .head(10)
pay_country = pay_country.rename(columns={'country':'Country Name'})
display(pay_country)

#Visualize the data
plt.figure(figsize=(10, 8))
b = sns.barplot(x='Total Amount of Payment', y='Country Name', data = pay_country)
b.set_title("Top 20 Total Amount of Sales by Country", size = 14)
b.set_xlabel('Total Amount of Payment', size = 14)
b.set_ylabel('Country Name', size = 14)
```

From the bar chart, we can conclude that **India** is the top customers in terms of payment amount. Then, it is followed by China, United States, Japan, and Mexico. **Interestingly**, the total amount of payment from Mexico (5th place) is almost half of India's total amount of payment (1st place).

## Answering Questions no. 3
**Visualize the contribution of payment sales by genre within the top 1 country in sales (India)**

```python
sg_country = df_all.groupby(['country','name'])['amount'].sum().reset_index(name='Payment Amount')
sg_country
```

```python
#Select data only from India.
sg_india = sg_country.loc[sg_country['country'] == 'India'] 
sg_india = sg_india.sort_values(by='Payment Amount', ascending = False) #Sort based on the highest total payment

#Visualize the data
plt.figure(figsize=(10, 8))
c = sns.barplot(x='Payment Amount', y='name', data = sg_india)
c.set_title("Total Amount of Sales by Genre in India", size = 14)
c.set_xlabel('Total Amount of Payment', size = 14)
c.set_ylabel('Genre', size = 14)
```

Based on the data, we can see **Sport** genre generated **the highest amount of payment in India**, followed by Documentary, Action, Animation, and Foreign.
